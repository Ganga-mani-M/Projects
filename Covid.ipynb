Forecasting the values of worldwide confirmed,death & recovered cases using prophet


```python
from prophet import Prophet
```

```python
confirmed = df.groupby('Date').sum()['Confirmed'].reset_index()
deaths = df.groupby('Date').sum()['Deaths'].reset_index()
recovered = df.groupby('Date').sum()['Recovered'].reset_index()
```

```python
path_output ='COVID-19 predictions using ML/Output_files_folder'
```

```python
confirmed.columns =['ds','y']
confirmed['ds'] =pd.to_datetime(confirmed['ds'])
x =open("COVID-19 predictions using ML/Output_files_folder/colab_op12.ipynb",'w')
print(confirmed.tail())
x.close()
```

Forcasting exmaple prophet
https://ecyy.medium.com/forecasting-by-fb-prophet-in-colab-c9d4db2d4195

```python
m = Prophet(interval_width= 0.95)
m.fit(confirmed)
future = m.make_future_dataframe(periods =7)
```

```python
future
```

```python
x =open("COVID-19 predictions using ML/Output_files_folder/colab_op13.ipynb",'w')
print(future.tail())
x.close()
```

```python
#predicting the future with date and upper and lower limit of y value
forecast =m.predict(future)
forecast 
```

```python
x =open("COVID-19 predictions using ML/Output_files_folder/colab_op14.ipynb",'w')
print(forecast[['ds','yhat','yhat_lower','yhat_upper']].tail(),file =x)
x.close()
```

```python
confirmed_forecast_plot = m.plot(forecast)
```

```python
confirmed_forecast_plot =m.plot_components(forecast)
```

Forecasting the worldwide Death using prophet (Base model)

```python

deaths.columns =['ds','y']
deaths['ds'] =pd.to_datetime(deaths['ds'])
m= Prophet(interval_width =0.95)
m.fit(deaths)
future =m.make_future_dataframe(periods =7)
print(future.tail())
forecaster =m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
deaths.forecast_plot = m.plot(forecast)
```

```python
deaths_forecast_plot= m.plot_components(forecast)
```

Forecasting the worldwide Recovered cases using prophet (Base model)

```python

recovered.columns =['ds','y']
recovered['ds'] =pd.to_datetime(recovered['ds'])
m= Prophet(interval_width =0.95)
m.fit(recovered)
future =m.make_future_dataframe(periods =7)
print(future.tail())
forecaster =m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
recovered_forecast_plot = m.plot(forecast)
```
